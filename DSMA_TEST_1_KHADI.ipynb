```python
data_transformed = pd.DataFrame()
data_transformed['log_sales_brand2'] = np.log(data['sales.brand2'])
data_transformed['log_feature_brand2'] = np.log(data['feature.brand2'] + 1)
data_transformed['log_display_brand2'] = np.log(data['display.brand2'] + 1)
data_transformed['log_assortment_brand2'] = np.log(data['assortment.brand2'])
data_transformed['log_price_brand2'] = np.log(data['price.brand2'])
data_transformed['log_tv_brand2'] = np.log(data['tv.brand2'] + 1)
data_transformed['log_digital_brand2'] = np.log(data['digital.brand2'] + 1)
data_transformed['log_ooh_brand2'] = np.log(data['ooh.brand2'] + 1)
data_transformed['log_magazine_brand2'] = np.log(data['magazine.brand2'] + 1)
```

```python
X = data_transformed[[
    'log_feature_brand2',
    'log_display_brand2',
    'log_assortment_brand2',
    'log_price_brand2',
    'log_tv_brand2',
    'log_digital_brand2',
    'log_ooh_brand2',
    'log_magazine_brand2',
]]
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(data_transformed['log_sales_brand2'], X).fit()
```

```python
model_summary = model.summary()
model_summary
```

Based on the regression results, the model explains approximately 64.1% of the variation in Pepsi's sales (𝑅2=0.641). This indicates a moderately strong relationship between the marketing mix variables and sales.
Among the predictors, feature advertising is statistically significant (𝑝<0.01) and has a large positive coefficient (β=1.024). This implies that a 1% increase in feature advertising is associated with an approximately 1.02% increase in Pepsi’s sales. Clearly, this variable is a major driver of sales growth.
Conversely, other variables such as display advertising, price, TV, and digital advertising are not statistically significant at the 5% level. Although the coefficient for price (β=−0.475) suggests a negative relationship with sales, the p-value of 0.097 indicates weaker evidence against the null hypothesis. Thus, price changes may have some effect, but this result should be interpreted cautiously.
In summary, the findings highlight the critical importance of feature advertising for Pepsi's sales performance, while other marketing mix elements show limited statistical evidence of impact in this model. Further investigation could explore interaction effects or nonlinear relationships.

```python
def adstock(series, decay):
    adstocked = np.zeros(len(series))
    adstocked[0] = series.iloc[0]
    for t in range(1, len(series)):
        adstocked[t] = series.iloc[t] + decay * adstocked[t-1]
    return adstocked
```

```python
lambda_decay = 0.6
data['adstock_tv'] = adstock(data['tv.brand2'], lambda_decay)
data['adstock_ooh'] = adstock(data['ooh.brand2'], lambda_decay)
data['adstock_magazine'] = adstock(data['magazine.brand2'], lambda_decay)
data['adstock_digital'] = adstock(data['digital.brand2'], lambda_decay)
```

```python
data_transformed = pd.DataFrame()
data_transformed['log_sales_brand2'] = np.log(data['sales.brand2'])
data_transformed['log_feature_brand2'] = np.log(data['feature.brand2'] + 1)
data_transformed['log_display_brand2'] = np.log(data['display.brand2'] + 1)
data_transformed['log_assortment_brand2'] = np.log(data['assortment.brand2'])
data_transformed['log_price_brand2'] = np.log(data['price.brand2'])
data_transformed['log_adstock_tv'] = np.log(data['adstock_tv'] + 1)
data_transformed['log_adstock_ooh'] = np.log(data['adstock_ooh'] + 1)
data_transformed['log_adstock_magazine'] = np.log(data['adstock_magazine'] + 1)
data_transformed['log_adstock_digital'] = np.log(data['adstock_digital'] + 1)
```

```python
X = data_transformed[[
    'log_feature_brand2',
    'log_display_brand2',
    'log_assortment_brand2',
    'log_price_brand2',
    'log_adstock_tv',
    'log_adstock_ooh',
    'log_adstock_magazine',
    'log_adstock_digital'
]]
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(data_transformed['log_sales_brand2'], X).fit()
```

```python
model.summary()
```

Based on the regression results, the model with adstock transformation explains approximately 70.8% of the variation in Pepsi's sales (R 2=0.708). This indicates a strong relationship between the marketing mix variables and sales when accounting for the longer-term effects of advertising.
Among the adstocked advertising variables, OOH  advertising (β=0.0124, p<0.01) and magazine advertising (β=0.0114, p<0.01) have statistically significant positive effects on sales. This suggests that a 1% increase in OOH or magazine adstock leads to a 0.012% and 0.011% increase in sales, respectively. These findings highlight the potential longer-term impact of these advertising channels on sales.
In contrast, TV (p=0.559) and digital (p=0.280) advertising adstock effects are not statistically significant, indicating limited evidence of their long-term impact on sales in this dataset.
For other marketing mix elements, feature advertising (β=0.9425, p<0.01) and assortment (β=0.7122, p<0.01) remain significant drivers of sales, while price (β=−1.1203, p<0.01) has a substantial negative effect.
In summary, the findings suggest that feature ads, assortment, and price play critical roles in driving sales, while OOH and magazine advertising exhibit significant longer-term effects. TV and digital advertising show limited evidence of impact in this model. Further exploration of interaction effects or alternate decay rates might provide additional insights.
